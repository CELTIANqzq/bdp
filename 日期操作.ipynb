```python
import pandas as pd
data = pd.read_csv("/Users/zhaozihao/Desktop/月日年.csv")
```

```python
data.info()
```

```python
import datetime
# 将str转换为datetime
time = datetime.datetime.strptime('1/1/2007', '%d/%m/%Y')
print(type(time))
print(data['publication_date'])
```

```python
# 转换为时间戳
for index,row in data.iterrows():
    print(row['publication_date'])
    print(index)
    row['publication_date'] = datetime.datetime.strptime(row['publication_date'] , '%m/%d/%Y')
    print(type(row['publication_date']))
```

```python
data['publication_date'] = pd.to_datetime(data['publication_date']).dt.year
columns = {'publication_date':'year'}
data.rename(columns=columns,inplace=True)
data['year'].apply(int).head()
```

```python
# 输出更改后的格式：月日年--》年
df = pd.DataFrame(data,columns = ["bookID","title","authors","average_rating","isbn","isbn13","language_code","num_pages","ratings_count","text_reviews_count","year","publisher"])
print(df)
df.to_csv("/Users/zhaozihao/Desktop/年.csv",index = False)
```
