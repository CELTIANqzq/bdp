```python
from sklearn import tree
from sklearn.datasets import load_wine
from sklearn.model_selection import train_test_split
```

```python
wine = load_wine()
```

```python
wine.data
```

```python
wine.target
```

```python
import pandas as pd
pd.concat([pd.DataFrame(wine.data),pd.DataFrame(wine.target)],axis=1)
```

```python
wine.feature_names
```

```python
wine.target_names
```

```python
# 分训练集测试集
Xtrain,Xtest,Ytrain,Ytest = train_test_split(wine.data,wine.target,test_size = 0.3)
```

```python
Xtrain.shape
```

```python
wine.data.shape
```

```python
clf = tree.DecisionTreeClassifier(criterion='entropy')
clf = clf.fit(Xtrain,Ytrain)
score = clf.score(Xtest,Ytest)
```

```python
score
```

```python
feature_name = ['酒精','苹果酸','灰','灰的碱性','镁','总酚','类黄酮','非黄烷类酚类',
                '花青素','颜色强度','色调','od280/od315稀释葡萄酒','脯氨酸']

import graphviz
dot_data = tree.export_graphviz(clf,
                                feature_names = feature_name,
                                class_names = ['琴酒','雪莉','贝尔摩德'],
                                filled=True,
                                rounded=True
                               )
graph = graphviz.Source(dot_data)
graph
```

```python
# 特征的重要性
clf.feature_importances_
```

```python
[*zip(feature_name,clf.feature_importances_)]
```

```python
# 不纯度是通过节点来计算的，决策树在建树时，是靠优化节点来追求一棵优化的树，但最优树的节点不可以保证最优的树
# 集成算法：建更多的不同的树，从中选取最好的
# 如何建不同的树？
# 在每次分支时，不使用全部特征，随机选取部分特征，从中选取不纯度相关指标最优的作为分支用的节点
# 就会产生问题：不稳定，score在变
# 解决方案 random_state = 30，用来设置分支中的随机模式的参数，在高维度时表现会更明显
clf = tree.DecisionTreeClassifier(criterion='entropy',random_state = 30)
clf = clf.fit(Xtrain,Ytrain)
score = clf.score(Xtest,Ytest)
score
```

```python
# splitter有两个参数值：random追求随机---树会变深，对训练集的拟合降低，best追求节点最优
clf = tree.DecisionTreeClassifier(criterion='entropy',random_state = 30,splitter='random')
clf = clf.fit(Xtrain,Ytrain)
score = clf.score(Xtest,Ytest)
score
```

```python
import graphviz
dot_data = tree.export_graphviz(clf,
                                feature_names = feature_name,
                                class_names = ['琴酒','雪莉','贝尔摩德'],
                                filled=True,
                                rounded=True
                               )
graph = graphviz.Source(dot_data)
graph
```

```python
# 检测训练结果
score_train = clf.score(Xtrain,Ytrain)
score_train
# 过拟合
```

```python
# 为了有更好的泛化能力，需要剪枝
clf = tree.DecisionTreeClassifier(criterion='entropy',
                                  random_state = 30,
                                  splitter='random',
                                  max_depth=3,
                                  min_samples_leaf=10,
#                                   min_samples_split=10
                                 )
clf = clf.fit(Xtrain,Ytrain)
dot_data = tree.export_graphviz(clf,
                                feature_names = feature_name,
                                class_names = ['琴酒','雪莉','贝尔摩德'],
                                filled=True,
                                rounded=True
                               )
graph = graphviz.Source(dot_data)
graph
```

```python
# 检测训练结果
score = clf.score(Xtest,Ytest)
score
```

```python
# 检测训练的拟合情况
score_train = clf.score(Xtrain,Ytrain)
score_train
```
