```python
import csv
import pandas as pd

path = '/Users/zhaozihao/Desktop/Dataset/'

# 转成utf-8的编码
with open(path + 'Train.csv', 'r', encoding='utf-8', errors='ignore') as infile, open(path + 'train_transcode.csv', 'w') as outfile:
     inputs = csv.reader(infile)
     output = csv.writer(outfile)
     output.writerow(["bookID","title","authors","average_rating","isbn","isbn13","language_code","num_pages","ratings_count","text_reviews_count","publication_date","publisher"])

     for index, row in enumerate(inputs):
         # 去除头部标签的内容
         if index == 0:
             continue
         output.writerow(row)

with open(path + 'Test.csv', 'r', encoding='utf-8', errors='ignore') as infile, open(path + 'test_transcode.csv', 'w') as outfile:
     inputs = csv.reader(infile)
     output = csv.writer(outfile)
     output.writerow(["bookID","title","authors","average_rating","isbn","isbn13","language_code","num_pages","ratings_count","text_reviews_count","publication_date","publisher"])
     
     for index, row in enumerate(inputs):
         # 去除头部标签的内容
         if index == 0:
             continue
         output.writerow(row)

train_file = "/Users/zhaozihao/Desktop/Dataset/train_transcode.csv"
test_file = "/Users/zhaozihao/Desktop/Dataset/test_transcode.csv"
train = pd.read_csv(train_file,encoding='utf-8')
test = pd.read_csv(test_file,encoding='utf-8')
```
