```python
# 判断评分在不在0-5之间
# 结果显示全部都在0-5
count_0 = 0
count_5 = 0
print('average_rating == 0的index为：')
for row in data.itertuples():
#     if(row.average_rating<=0 and row.average_rating>5):
    if(row.average_rating == 0):
        print(row.bookID)
        count_0 = count_0 + 1
print('-------------合计有：')
print(count_0)
print('average_rating == 5的index为：')
for row in data.itertuples():
    if(row.average_rating == 5):
        print(row.bookID)
        count_5 = count_5 + 1
print('-------------合计有：')
print(count_5)
```

```python
#判断num_pages等于0的index
count_num_pages = 0
for row in data.itertuples():
    if row.num_pages == 0 :
        print(row.bookID)
        count_num_pages = count_num_pages + 1
print('-------------合计有:')
print(count_num_pages)
```

```python
# 将num_pages==0的数据删除
# 遇到问题：多出第一列
import csv

with open('/Users/zhaozihao/Desktop/年.csv', 'r') as fin, open('/Users/zhaozihao/Desktop/删除问题数据结果.csv', 'w', newline='') as fout:

    # define reader and writer objects
    reader = csv.reader(fin, skipinitialspace=True)
    writer = csv.writer(fout, delimiter=',')

    # write headers
    writer.writerow(next(reader))

    # iterate and write rows based on condition
    for row in data.itertuples():
        if row.num_pages == 0 :
            continue
        else:
            writer.writerow(row)
```

```python
# 解决问题：删除第一列
import pandas as pd
import csv
 
data = pd.read_csv("/Users/zhaozihao/Desktop/删除问题数据结果.csv")
# 去除上述步骤的第一列数据 
data_new=data.drop(columns=["ID"]) 
# 写入新的csv文件 
data_new.to_csv("/Users/zhaozihao/Desktop/简洁.csv",index=0)
```
